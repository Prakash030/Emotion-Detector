```python
Classes = ["0","1","2","3","4","5","6"]  ##list of classes(emotion level)
```

Tranfer learning uses 224x224 size of an image so all the images are converted into the size 224x224

# Read all images and convert them to array

```python
training_Data = []  ## data array
def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category)  ## label 0,1,2...
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array = cv2.resize(img_array,(img_size,img_size))
                training_Data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

```python
batch_size = 512
```

```python
num_batches = len(training_Data) // batch_size
```

```python
for i in range(num_batches):
    start = i * batch_size
    end = start + batch_size
    batch_data = training_Data[start:end]
```

```python
print(len(batch_data))
```

```python
import random

random.shuffle(batch_data)
```

```python
X = []  ## data / feature
y=[]  ## label

for features,label in batch_data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1, img_size, img_size, 3)  ## converting it to 4 dimension
```

```python
X.shape
```

```python
X = X/255.0;  ## we are normalizing it
```

```python
y[52]
```

```python
Y = np.array(y)
Y.shape
```

# Deep Learnng model for training - Transfer Learning

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2()  ## Pretrained mode
```

```python
model.summary()
```

# Transfer learning - Tuning, weights will start from last check point

```python
base_input = model.layers[0].input
```

```python
base_output = model.layers[-2].output
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output)  ## adding new layer, after the output of 
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(64)(final_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7,activation='softmax')(final_output)  ## total classes are 7
```

```python
final_output
```

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
new_model.fit(X,Y, epochs = 15, batch_size=30)
```

```python
new_model.save('Final_model_95p.h5')
```
